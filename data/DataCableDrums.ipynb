```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
forecastdrum = pd.read_csv("drum_1167_linear_forecast.csv")
forecastdrum['date'] = pd.to_datetime(forecastdrum['date'])
```

```python
forecastdrum.head()
```

```python
# Sort by date (just in case)
forecastdrum = forecastdrum.sort_values('date')

plt.figure(figsize=(12, 6))

# Plot lines
plt.plot(forecastdrum['date'], forecastdrum['daily_min_cable_length_m'], label='Min')
plt.plot(forecastdrum['date'], forecastdrum['daily_avg_cable_length_m'], label='Average')
plt.plot(forecastdrum['date'], forecastdrum['daily_max_cable_length_m'], label='Max')
plt.plot(forecastdrum['date'], forecastdrum['linear_forecast_m'], label='Forecast')

# Optional: fill area between min and max for better visualization
plt.fill_between(
    forecastdrum['date'],
    forecastdrum['daily_min_cable_length_m'],
    forecastdrum['daily_max_cable_length_m'],
    alpha=0.2
)

# Labels & styling
plt.xlabel('Date')
plt.ylabel('Cable Length (m)')
plt.title('Daily Cable Length Statistics of 1167')
plt.legend()
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
