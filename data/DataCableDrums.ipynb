```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
forecastdrum = pd.read_csv("drum_1167_linear_forecast.csv")
forecastdrum['date'] = pd.to_datetime(forecastdrum['date'])
```

```python
forecastdrum.head()
```

```python
# Sort by date (just in case)
forecastdrum = forecastdrum.sort_values('date')

# Use Arial for text rendering
plt.rcParams['font.family'] = 'Arial'

plt.figure(figsize=(12, 6))

# Plot lines
plt.plot(forecastdrum['date'], forecastdrum['daily_avg_cable_length_m'], label='Reale Kabellänge', color='#7f8c8d', linewidth=1.8)
plt.plot(forecastdrum['date'], forecastdrum['linear_forecast_m'], label='Vorhersage Kabellänge', color='#ff7100', linewidth=1.8)

# Highlight all periods where Durchschnitt is below Vorhersage
below_forecast = forecastdrum['daily_avg_cable_length_m'] < forecastdrum['linear_forecast_m']
plt.fill_between(
    forecastdrum['date'],
    forecastdrum['daily_avg_cable_length_m'],
    forecastdrum['linear_forecast_m'],
    where=below_forecast,
    color="#ff0004",
    alpha=0.25,
    interpolate=True,
    label='reale unter vorhergesagter Kabellänge'
 )

# If available, draw the reorder threshold
if 'order_threshold_m' in forecastdrum.columns:
    threshold_value = forecastdrum['order_threshold_m'].iloc[-1]
    plt.axhline(
        y=threshold_value,
        color='#4e3d30',
        linestyle='--',
        linewidth=2.0,
        label='Bestellschwelle'
    )


# Labels & styling
plt.xlabel('Datum', fontname='Arial')
plt.ylabel('Kabel Länge (m)', fontname='Arial')
plt.title('Tägliche Kabellänge und Vorhersage: Trommel 1167', fontname='Arial')
plt.grid(True, alpha=0.3)
plt.legend(loc='lower left')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python

# Polynomial Regression (Grad 2) als Beispiel der neuen Logik
x = np.arange(len(forecastdrum))
y = forecastdrum['daily_avg_cable_length_m'].values
coeffs = np.polyfit(x, y, 4)
poly = np.poly1d(coeffs)
forecastdrum['poly_forecast_m'] = poly(x)

# Farbe/Schrift wie in deinem aktuellen Stil
plt.rcParams['font.family'] = 'Arial'
plt.figure(figsize=(12, 6))

# Reale Kabellänge
plt.plot(
    forecastdrum['date'],
    forecastdrum['daily_avg_cable_length_m'],
    label='Reale Kabellänge',
    color='#7f8c8d',
    linewidth=1.8
)

# Bisherige lineare Vorhersage
plt.plot(
    forecastdrum['date'],
    forecastdrum['linear_forecast_m'],
    label='Lineare Vorhersage',
    color='#ff7100',
    linewidth=1.8,
    alpha=0.75
)

# Neue polynomiale Vorhersage
plt.plot(
    forecastdrum['date'],
    forecastdrum['poly_forecast_m'],
    label='Polynomiale Vorhersage (Grad 2)',
    color='#ff7100',
    linewidth=3.0
)

# Unterschied linear vs polynomial visualisieren
delta = (forecastdrum['poly_forecast_m'] - forecastdrum['linear_forecast_m']).abs()
plt.fill_between(
    forecastdrum['date'],
    forecastdrum['linear_forecast_m'],
    forecastdrum['poly_forecast_m'],
    color='#ff0004',
    alpha=0.20,
    label='Differenz linear vs polynomial'
 )

# Bestellschwelle falls vorhanden
if 'order_threshold_m' in forecastdrum.columns:
    threshold_value = forecastdrum['order_threshold_m'].iloc[-1]
    plt.axhline(
        y=threshold_value,
        color='#4e3d30',
        linestyle='--',
        linewidth=2.0,
        label='Bestellschwelle'
    )

# KPI-Textbox fuer den Kundentermin
mean_delta = float(delta.mean())
max_delta = float(delta.max())
plt.text(
    0.01, 0.98,
    f'Ø Differenz: {mean_delta:.2f} m | Max Differenz: {max_delta:.2f} m',
    transform=plt.gca().transAxes,
    va='top',
    ha='left',
    fontsize=10,
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='#dddddd')
 )

plt.xlabel('Datum', fontname='Arial')
plt.ylabel('Kabel Länge (m)', fontname='Arial')
plt.title('Vergleich: Lineare vs. polynomiale Vorhersage (Trommel 1167)', fontname='Arial')
plt.grid(True, alpha=0.3)
plt.legend(loc='lower left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
